### A key challenge for property sellers is to determine the sale price of the property. The ability to predict the exact property value is beneficial for property investors as well as for buyers to plan their finances according to the price trend. The property prices depend on the number of features like the property area, basement square footage, year built, number of bedrooms, and so on

## Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

## Setting pandas options for convinient display

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
np.set_printoptions(suppress=False)
```

## Importing and Reading the Data

```python
df=pd.read_csv('HousePrices.csv')
df.head(20)
```

```python
df.shape
```

```python
df_manipulated=df.copy(deep=True)
df_manipulated.drop(['Id'],axis=1,inplace=True)
```

```python
df_manipulated.isnull().sum()
```

```python
df_manipulated.drop(['Alley','FireplaceQu','PoolQC','Fence','MiscFeature','LotFrontage','MasVnrType','GarageType','GarageYrBlt','GarageFinish','GarageType'],axis=1,inplace=True)
```

```python
df_manipulated.fillna(df_manipulated.mean(numeric_only=True),inplace=True)
```

```python
df_manipulated.dropna(inplace=True)
```

```python
df_model=df_manipulated.select_dtypes(exclude=['object'])
```

# Modelling

```python
X=df_model.drop('Property_Sale_Price',axis=1)
y=df_model.Property_Sale_Price
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
reg=linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```
